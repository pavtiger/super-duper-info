````python
import pandas as pd
````

...и вот таким нехитрым кодом сразу импортируем и отображаум нашёу таблицу. Круто, да?

````python
data = pd.read_csv(path)
data
````

Сразу покажу крутой метод. (Нам он не особо пригодится, по крайней мере вот прямо сейчас, но он круто!)

````python
data.describe()
````

Поудаляем всякие ненужные столбцы.

````python
data = data[[\
      'age', 'sex', 'city', 'province', 'country', 'wuhan(0)_not_wuhan(1)', 'latitude',	'longitude', 'geo_resolution', \
      'date_onset_symptoms', 'date_admission_hospital', 'date_confirmation', 'symptoms', 'lives_in_Wuhan'\
      ]]
data
````

После этого я решил отсортировать все случаи по дате подтверждения вируса.

````python
data.sort_values(by='date_confirmation')
````

...но тут меня ждал облом.

Части дат не хватает. Как бы с этим поработать?
Вообще, pandas умеет вот в такую интересную индексацию.

````python
# Вот какой-то столбик
data['country']
````

````python
# И столбик превращается...
data['country'] == 'Russia'
````

````python
# Столбик превращается...
data[data['country'] == 'Russia']
````

Таким образом, мы понимаем, что в качестве индекса в таблицу можно подставить массив булевых значений.
Подобного результата можно добиться специальными методами. Угадайте по названию этого, что он делает :)

(False не видно, но они есть!)

````python
data['date_confirmation'].notna()
````

А теперь по старой схеме удалим плохие строки.

````python
data = data[data['date_confirmation'].notna()]
````

Теперь отсортируем!

````python
data.sort_values(by='date_confirmation')
````

Вроде всё норм, строки с NaN исчезли. Но сортировка всё равно какая-то не очень. Почему так? Так происходит из-за того, что строки отсортировались по алфавиту. Давайте попробуем привести эти значения к другому типу.
```pd.to_datetime()``` поможет нам в этом деле.

````python
# она даже, вроде как, работает
pd.to_datetime(["01.01.2001", "01.01.2001"], format="%d.%m.%Y")
````

Приведём к нужному типу весь столбец...

````python
pd.to_datetime(data["date_confirmation"],  format="%d.%m.%Y")
````

Ой, что-то пошло не так. Но что? Поищем...

````python
for dt in data['date_confirmation']:
    try:
        pd.to_datetime(dt,  format="%d.%m.%Y")
    except ValueError:
        print("Вот он, гад!", dt)
````

Ну, ок, возьмём и уточним, чтобы проще было.

````python
mask = data['date_confirmation'] == "25.02.2020 - 26.02.2020"
data.loc[mask, 'date_confirmation'] = "25.02.2020"
data
# Я пока не разобрался, как поправить то, на что он ругается. Но у меня вроде работает.
````

И вот таким брутальным приёмом сохраним изменение.

````python
data["date_confirmation"] = pd.to_datetime(data["date_confirmation"],  format="%d.%m.%Y")
data
````

````python
data.sort_values(by='date_confirmation') 
# обратите внимание, что этот код не сотрирует саму таблицу, а только возвращает отсортированную!
````

````python
data = data.sort_values(by='date_confirmation') 
````

````python
from matplotlib import pyplot as plt
````

````python
plt.plot(data['date_confirmation'])
plt.show()
````
