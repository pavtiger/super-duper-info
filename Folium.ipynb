```python
import pandas as pd

path = 'kek.csv'
data = pd.read_csv(path)
data
```

```python
import random

m = folium.Map(
    location=[45.372, -121.6972],
    zoom_start=12,
    tiles='Stamen Terrain'
)

for i in range(1000):
    j = random.randint(0, len(data.ID))
    tooltip = data.age[j]
    folium.Marker([data.latitude[j], data.longitude[j]], popup='<i>i am ill</i>', tooltip=tooltip).add_to(m)

m
```
